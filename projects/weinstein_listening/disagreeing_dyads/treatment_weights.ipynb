```python
learningOrPersuasionList = ["LG", "PG"]
topicList = [
    "bathroom_choice",
    "censoring_extremism",
    "education_employability",
    "privatise_NHS",
    "raise_cost_of_uni",
    "remote_classes",
    "restricting_immigrants",
    "restricting_legal_abortion"
]
p0List = [
    "1",
    "2",
    "3",
    "4",
    "5",
    "6",
    "7",
]

p1List = [
    "1",
    "2",
    "3",
    "4",
    "5",
    "6",
    "7",
]
```

```python
import re

def extract_digit(s):
    match = re.search(r'\d', s)
    if match:
        return int(match.group())
    

conditions = []
weights = []
for learningOrPersuasion in learningOrPersuasionList:
    for topic in topicList:
        for p0 in p0List:
            for p1 in p1List:
                lowerPosNumber = extract_digit(p0)
                upperPosNumber = extract_digit(p1)
                if abs(lowerPosNumber - upperPosNumber) >= 2:
                    conditions.append(
                        learningOrPersuasion + "_condition_" + topic + "_" + p0 + "_" + p1)
                    weights.append(abs(upperPosNumber - lowerPosNumber))

print(len(conditions))
```
