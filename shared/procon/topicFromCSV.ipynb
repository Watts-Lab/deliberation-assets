```python
import pandas as pd
```

```python
questions = pd.read_csv("questions.csv", delimiter="|")
questions.columns = [col.strip() for col in questions.columns]

def makeShortTitle(row):
    try:
        if row['url'].split("//")[1].startswith("procon"):
            return row["title"].strip().replace(" ", "").replace(".", "").replace("-","").replace("(","").replace(")","").replace("&","").lower()
        else:
            return row["url"].split("//")[1].split(".")[0]
    except:
        print(row)
    
questions["short_title"] = questions.apply(makeShortTitle, axis=1)
assert questions['short_title'].value_counts().max() == 1

pd.set_option('display.max_rows', None)
questions

```

```python
def makeSurvey(row):

    output_text = f"""---
name: shared/procon/{row["short_title"]}_survey.md
title: {row["title"]}
type: multipleChoice
author: reprinted with permission of ProCon.org
url: {row["url"]}
---

### {row["question"]}

My opinion is...

---

- Strongly Yes
- Moderately Yes
- Leaning towards Yes
- Neutral
- Leaning towards No
- Moderately No
- Strongly No

"""

    with open(f"{row['short_title']}_survey.md", "w") as f:
        f.write(output_text)

questions.apply(makeSurvey, axis=1)
```

```python
def makeDiscussionPrompt(row):

    output_text = f"""---
name: shared/procon/{row["short_title"]}_topic.md
title: {row["title"]}
type: noResponse
author: reprinted with permission of ProCon.org
url: {row["url"]}
---

### {row["question"]}

---

"""

    with open(f"{row['short_title']}_topic.md", "w") as f:
        f.write(output_text)

questions.apply(makeDiscussionPrompt, axis=1)
```
